## Problem 1

Plot the function and it's derivative

Plot of function 3x-6x^3-x^4:

```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(-6, 1, 100) 
fig = plt.figure() 
plt.plot(x, 3*x-6*x**3-x**4) 
plt.show() 
```

Derivative of function:

```python
x = np.linspace(-6, 1, 100) 
fig = plt.figure() 
plt.plot(x, -4*x**3-18*x**2+3) 
plt.show() 
```

Approximate the x values where the function is at a local max or min

```python
import torch
x_point = -6.0
learning_rate = 0.005
coordinates = []
while x_point >= -6.0 and x_point <= 1.0:
    x = torch.tensor(x_point, requires_grad = True)
    y = 3*x-6*x**3-x**4
    y.backward()
    if abs(x.grad.numpy()) <= 0.01:
        print(x_point, x.grad.numpy())
        coordinates.append(x_point)
        x_point = x_point + 0.005
    x_point = x_point + abs(learning_rate*x.grad.numpy())
```

```python
print(coordinates)
```

Plot the max and min as points on the function

```python
import numpy as np 
import matplotlib.pyplot as plt 
x = np.linspace(-6, 1, 100) 
fig = plt.figure() 
plt.plot(x, 3*x-6*x**3-x**4)
coordinates = np.array(coordinates)
plt.scatter(coordinates, 3*coordinates-6*coordinates**3-coordinates**4)
plt.title('y = 3^x-6x^3-x^4+4')
plt.grid(b=True)
plt.show() 
```

Plot the function and it's derivatives

3D plot of function cos(x)*sin(y):

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  
# Axes3D import has side effects, it enables using projection='3d' in add_subplot
import matplotlib.pyplot as plt
import random

def fun(x, y):
    return np.cos(x)*np.sin(y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-np.pi, np.pi, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array(fun(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
